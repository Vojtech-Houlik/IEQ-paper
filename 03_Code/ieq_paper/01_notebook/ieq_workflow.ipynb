# IEQ Workflow

Single reproducible notebook for the overall IEQ paper.


## 1. Setup

Imports, paths, constants, and plotting style.

## 2. Dataset Definition and Cleaning

Load raw data, select target and features, create the clean Excel dataset.

## 3. Missingness and Target Distribution

Create completeness plots, complete-case summaries, and target distributions.

## 4. Target Simplification

Create dissatisfied / neutral / satisfied target.

## 5. Model Comparison

Linear or ordinal baseline, Random Forest, XGBoost/gradient boosting, and ANN.

## 6. Best Model and Feature Importance

Select best model, generate feature importance, and write interpretation notes.

## 7. External Dataset and Transfer

Dataset compatibility, feature alignment, external validation, and transfer learning experiments.
