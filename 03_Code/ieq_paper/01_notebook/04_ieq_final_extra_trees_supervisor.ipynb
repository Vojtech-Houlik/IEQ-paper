# IEQ Final Extra Trees Model

This notebook trains and evaluates the final paper-facing IEQ model using the model-ready dataset. It does not clean data, impute missing values, or search for new hyperparameters. The Extra Trees hyperparameters are fixed to the selected final configuration used for the article.

## 1. Imports and Reproducibility Settings

```python
from __future__ import annotations

import platform
import sys
import time
from pathlib import Path

import matplotlib
if "ipykernel" not in sys.modules:
    matplotlib.use("Agg")
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import shap
import sklearn
from sklearn.base import clone
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import accuracy_score, balanced_accuracy_score, confusion_matrix, f1_score, mean_absolute_error
from sklearn.model_selection import StratifiedKFold, StratifiedShuffleSplit
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

try:
    from IPython.display import display # type: ignore
except ModuleNotFoundError:
    def display(value):
        print(value)

SEED = 20260507
N_SPLITS = 5
TARGET = "IEQ satisfaction 3-class"
DROP_FROM_FEATURES = [TARGET, "TimeVote"]
CLASS_ORDER = ["dissatisfied", "neutral", "satisfied"]
CLASS_TO_NUMBER = {label: index for index, label in enumerate(CLASS_ORDER)}
PERFORMANCE_METRICS = ["accuracy", "balanced_accuracy", "macro_f1", "weighted_f1", "ordinal_mae"]
DOMAIN_IMPORTANCE_REPEATS = 5
RAW_IMPORTANCE_TOP_N = 15
SHAP_SAMPLE_SIZE = 1000
FEATURE_CATEGORIES = {
    "Thermal environment": ["Temperature", "RH", "CLO", "Ttrend", "AC on/off"],
    "Indoor air quality": ["CO2"],
    "Lighting": ["Lighting"],
    "Acoustic": ["Sound"],
    "Occupant / schedule": ["Student", "Age", "Gender", "EA", "Vote hour", "Vote weekday", "Moment"],
    "Classroom position": ["LocationBack", "LocationFront", "LocationLeft", "LocationRight", "LocationMiddle"],
}
PLOT_COLORS = {
    "final_model": "#2A9D8F",
    "reference_model": "#B8C0C8",
    "text": "#111827",
    "grid": "#D9DEE3",
}

plt.rcParams.update(
    {
        "figure.facecolor": "white",
        "axes.facecolor": "white",
        "savefig.facecolor": "white",
        "font.family": "sans-serif",
        "font.size": 10.5,
        "axes.edgecolor": "#AAB2B8",
        "axes.grid": True,
        "grid.color": PLOT_COLORS["grid"],
        "grid.alpha": 0.85,
    }
)


def show_or_close(fig) -> None:
    if "ipykernel" in sys.modules:
        plt.show()
    else:
        plt.close(fig)

FINAL_EXTRA_TREES_PARAMS = {
    "n_estimators": 700,
    "criterion": "entropy",
    "bootstrap": True,
    "max_samples": 0.9,
    "max_features": None,
    "max_depth": 20,
    "min_samples_leaf": 2,
    "min_samples_split": 10,
    "class_weight": "balanced",
    "n_jobs": 1,
    "random_state": SEED,
}

print("Python:", platform.python_version())
print("pandas:", pd.__version__)
print("numpy:", np.__version__)
print("scikit-learn:", sklearn.__version__)
print("Random seed:", SEED)
```

## 2. Locate and Load the Model-Ready Dataset

The notebook first looks for the original Prism folder structure. If the notebook is shared separately, placing `ieq_model_dataset.xlsx` in the same working folder is enough.

```python
DATASET_FILE_NAME = "ieq_model_dataset.xlsx"
DATASET_RELATIVE_PATH = Path("02_Datasets") / "model_ready" / DATASET_FILE_NAME


def find_dataset_path() -> Path:
    candidates = []
    for root in [Path.cwd(), *Path.cwd().parents]:
        candidates.extend(
            [
                root / DATASET_RELATIVE_PATH,
                root / "Prism" / DATASET_RELATIVE_PATH,
                root / DATASET_FILE_NAME,
            ]
        )
    for candidate in candidates:
        if candidate.exists():
            return candidate.resolve()
    raise FileNotFoundError(
        "Could not find ieq_model_dataset.xlsx. Keep the Prism folder structure, "
        "or place ieq_model_dataset.xlsx in the current working folder."
    )


DATASET_PATH = find_dataset_path()
if DATASET_PATH.parent.name == "model_ready" and DATASET_PATH.parent.parent.name == "02_Datasets":
    PROJECT_ROOT = DATASET_PATH.parent.parent.parent
    OUTPUT_DIR = PROJECT_ROOT / "03_Code" / "ieq_paper" / "01_notebook" / "ieq_final_model_outputs"
else:
    OUTPUT_DIR = Path.cwd() / "ieq_final_model_outputs"
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

data = pd.read_excel(DATASET_PATH, sheet_name="data")

print("Dataset:", DATASET_PATH)
print("Rows, columns:", data.shape)
display(data.head())
```

## 3. Select Predictors

`IEQ satisfaction 3-class` is the target and must not be used as a predictor. `TimeVote` is the original timestamp metadata column, so the model uses the derived time-context inputs `Vote hour` and `Vote weekday` instead.

```python
feature_columns = [column for column in data.columns if column not in DROP_FROM_FEATURES]

X = data[feature_columns].copy()
y = data[TARGET].astype(str)

numeric_features = X.select_dtypes(include=["number"]).columns.tolist()
categorical_features = [column for column in X.columns if column not in numeric_features]

print("Target:", TARGET)
print("Number of predictors:", len(feature_columns))
print("Numeric predictors:", numeric_features)
print("Categorical / boolean predictors:", categorical_features)
print("Class distribution:")
display(y.value_counts().reindex(CLASS_ORDER).rename("rows").to_frame())

category_feature_map = {
    category: [feature for feature in features if feature in feature_columns]
    for category, features in FEATURE_CATEGORIES.items()
}
category_feature_map = {category: features for category, features in category_feature_map.items() if features}
domain_feature_map = {feature: [feature] for feature in feature_columns}
domain_feature_table = pd.DataFrame(
    [
        {"domain": category, "n_features": len(features), "features": ", ".join(features)}
        for category, features in category_feature_map.items()
    ]
)
with pd.option_context("display.max_colwidth", None, "display.max_rows", None):
    display(domain_feature_table)
```

## 4. Final Model Pipeline

The preprocessing is fitted only inside each training fold during cross-validation. Numeric predictors are standardized and categorical or boolean predictors are one-hot encoded. The classifier uses the fixed final Extra Trees hyperparameters.

```python
def make_one_hot_encoder():
    try:
        return OneHotEncoder(handle_unknown="ignore", sparse_output=False)
    except TypeError:
        return OneHotEncoder(handle_unknown="ignore", sparse=False)


def make_preprocessor():
    return ColumnTransformer(
        transformers=[
            ("numeric", StandardScaler(), numeric_features),
            ("categorical", make_one_hot_encoder(), categorical_features),
        ],
        remainder="drop",
    )


def make_final_model():
    return ExtraTreesClassifier(**FINAL_EXTRA_TREES_PARAMS)


def make_final_pipeline():
    return Pipeline(
        steps=[
            ("preprocess", make_preprocessor()),
            ("model", make_final_model()),
        ]
    )


display(pd.Series(FINAL_EXTRA_TREES_PARAMS, name="value").to_frame())
```

## 5. Deterministic 5-Fold Cross-Validation

This section evaluates the fixed final model. No hyperparameter search is performed.

```python
def ordinal_mae(y_true, y_pred):
    true_numbers = pd.Series(y_true).map(CLASS_TO_NUMBER).to_numpy()
    pred_numbers = pd.Series(y_pred).map(CLASS_TO_NUMBER).to_numpy()
    return float(mean_absolute_error(true_numbers, pred_numbers))


def evaluate_predictions(y_true, y_pred):
    return {
        "accuracy": float(accuracy_score(y_true, y_pred)),
        "balanced_accuracy": float(balanced_accuracy_score(y_true, y_pred)),
        "macro_f1": float(f1_score(y_true, y_pred, average="macro", zero_division=0)),
        "weighted_f1": float(f1_score(y_true, y_pred, average="weighted", zero_division=0)),
        "ordinal_mae": float(ordinal_mae(y_true, y_pred)),
    }


def grouped_domain_permutation_rows(pipeline, X_test, y_test, baseline_macro_f1, fold):
    rows = []
    rng = np.random.default_rng(SEED + fold * 1000)
    for domain, domain_features in domain_feature_map.items():
        feature_list = ", ".join(domain_features)
        for repeat in range(1, DOMAIN_IMPORTANCE_REPEATS + 1):
            shuffled_positions = rng.permutation(len(X_test))
            X_permuted = X_test.copy()
            for feature in domain_features:
                X_permuted[feature] = X_permuted[feature].iloc[shuffled_positions].to_numpy()
            permuted_pred = pipeline.predict(X_permuted)
            permuted_macro_f1 = float(f1_score(y_test, permuted_pred, average="macro", zero_division=0))
            rows.append(
                {
                    "fold": fold,
                    "repeat": repeat,
                    "domain": domain,
                    "features": feature_list,
                    "baseline_macro_f1": baseline_macro_f1,
                    "permuted_macro_f1": permuted_macro_f1,
                    "macro_f1_drop": baseline_macro_f1 - permuted_macro_f1,
                }
            )
    return rows


cv = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=SEED)
fold_rows = []
prediction_rows = []
domain_importance_rows = []

for fold, (train_index, test_index) in enumerate(cv.split(X, y), start=1):
    pipeline = make_final_pipeline()
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    start = time.perf_counter()
    pipeline.fit(X_train, y_train)
    fit_seconds = time.perf_counter() - start

    start = time.perf_counter()
    y_pred = pipeline.predict(X_test)
    predict_seconds = time.perf_counter() - start

    metrics = evaluate_predictions(y_test, y_pred)
    domain_importance_rows.extend(grouped_domain_permutation_rows(pipeline, X_test, y_test, metrics["macro_f1"], fold))
    fold_rows.append(
        {
            "fold": fold,
            "train_rows": len(train_index),
            "test_rows": len(test_index),
            **metrics,
            "fit_seconds": fit_seconds,
            "predict_seconds": predict_seconds,
            "total_seconds": fit_seconds + predict_seconds,
        }
    )

    for source_index, true_label, predicted_label in zip(test_index, y_test, y_pred):
        prediction_rows.append(
            {
                "source_index": int(source_index),
                "fold": fold,
                "y_true": true_label,
                "y_pred": predicted_label,
            }
        )

fold_results = pd.DataFrame(fold_rows)
cv_predictions = pd.DataFrame(prediction_rows)
domain_importance = pd.DataFrame(domain_importance_rows)
domain_importance_summary = (
    domain_importance
    .groupby(["domain", "features"], as_index=False)
    .agg(
        importance_mean=("macro_f1_drop", "mean"),
        importance_std=("macro_f1_drop", "std"),
        baseline_macro_f1_mean=("baseline_macro_f1", "mean"),
        permuted_macro_f1_mean=("permuted_macro_f1", "mean"),
    )
    .sort_values("importance_mean", ascending=False)
    .reset_index(drop=True)
)

summary = fold_results[PERFORMANCE_METRICS].agg(["mean", "std"]).T.reset_index(names="metric")
fold_metrics = fold_results[["fold", "train_rows", "test_rows", *PERFORMANCE_METRICS]].copy()

display(summary)
```

## 6. Confusion Matrix and Saved Outputs

```python
confusion = pd.DataFrame(
    confusion_matrix(cv_predictions["y_true"], cv_predictions["y_pred"], labels=CLASS_ORDER),
    index=[f"true_{label}" for label in CLASS_ORDER],
    columns=[f"pred_{label}" for label in CLASS_ORDER],
)

fold_metrics.to_csv(OUTPUT_DIR / "final_extra_trees_5fold_folds.csv", index=False)
summary.to_csv(OUTPUT_DIR / "final_extra_trees_5fold_summary.csv", index=False)
cv_predictions.to_csv(OUTPUT_DIR / "final_extra_trees_5fold_predictions.csv", index=False)
confusion.to_csv(OUTPUT_DIR / "final_extra_trees_confusion_matrix.csv")
domain_importance.to_csv(OUTPUT_DIR / "final_extra_trees_domain_importance_folds.csv", index=False)
domain_importance_summary.to_csv(OUTPUT_DIR / "final_extra_trees_domain_importance_summary.csv", index=False)

display(confusion)

confusion_percent = confusion.div(confusion.sum(axis=1), axis=0) * 100
fig, ax = plt.subplots(figsize=(6.2, 5.0))
image = ax.imshow(confusion_percent, cmap="Blues", vmin=0, vmax=100)

ax.set_title("Final Extra Trees confusion matrix")
ax.set_xlabel("Predicted class")
ax.set_ylabel("True class")
ax.set_xticks(range(len(CLASS_ORDER)))
ax.set_xticklabels(CLASS_ORDER, rotation=25, ha="right")
ax.set_yticks(range(len(CLASS_ORDER)))
ax.set_yticklabels(CLASS_ORDER)

for row_index, true_label in enumerate(CLASS_ORDER):
    for column_index, pred_label in enumerate(CLASS_ORDER):
        count = confusion.iloc[row_index, column_index]
        percent = confusion_percent.iloc[row_index, column_index]
        text_color = "white" if percent >= 50 else PLOT_COLORS["text"] # type: ignore
        ax.text(column_index, row_index, f"{count}\n{percent:.1f}%", ha="center", va="center", color=text_color)

colorbar = fig.colorbar(image, ax=ax, fraction=0.046, pad=0.04)
colorbar.set_label("Row percentage")
fig.tight_layout()
fig.savefig(OUTPUT_DIR / "final_extra_trees_confusion_matrix.png", dpi=300, bbox_inches="tight")
fig.savefig(OUTPUT_DIR / "final_extra_trees_confusion_matrix.pdf", bbox_inches="tight")
show_or_close(fig)

print("Saved outputs to:", OUTPUT_DIR.resolve())
```

## 9. Raw Domain and Feature Importance

This section reports raw-domain permutation importance for the final Extra Trees model using the same stratified 5-fold cross-validation as the performance evaluation. In each held-out fold, the trained fold model is first scored normally with macro F1. Then one model-ready predictor is randomly shuffled within that held-out fold while all other predictors stay unchanged, and macro F1 is recomputed. This permutation is repeated five times per fold. Importance is the mean decrease in held-out macro F1 after shuffling; larger positive values indicate predictors the model relied on more. The plotted bars show these means only; fold/repeat variability remains available in the CSV output but is not drawn in the figure.

For article-facing comparison, the final figure also places three complementary importance estimates side by side for the Extra Trees model: out-of-fold permutation importance, built-in impurity-based Extra Trees importance, and mean absolute TreeSHAP importance. The three measures use different natural units, so the comparison plot normalizes each method to its own maximum while keeping original-scale labels on the bars. TreeSHAP is computed on a deterministic stratified sample of 1,000 rows for tractability.

```python
domain_importance.to_csv(OUTPUT_DIR / "final_extra_trees_domain_importance_folds.csv", index=False)
domain_importance_summary.to_csv(OUTPUT_DIR / "final_extra_trees_domain_importance_summary.csv", index=False)
top_raw_domain_importance = domain_importance_summary.head(RAW_IMPORTANCE_TOP_N).copy()

plot_data = top_raw_domain_importance.sort_values("importance_mean", ascending=True)
fig, ax = plt.subplots(figsize=(7.2, 6.2))
ax.barh(plot_data["domain"], plot_data["importance_mean"], color=PLOT_COLORS["final_model"], edgecolor="#6B7280", linewidth=0.45)
ax.set_title("Final Extra Trees raw-domain importance")
ax.set_xlabel("Mean held-out macro F1 decrease after permutation")
ax.grid(False)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
for value, domain in zip(plot_data["importance_mean"], plot_data["domain"]):
    ax.text(value + max(plot_data["importance_mean"].max(), 0.001) * 0.015, domain, f"{value:.3f}", va="center", fontsize=8.5, color=PLOT_COLORS["text"])
fig.text(0.5, -0.02, "Permutation importance: each bar is the mean macro-F1 drop after shuffling one model-ready predictor in held-out CV folds.", ha="center", color=PLOT_COLORS["text"], fontsize=8.5)
fig.tight_layout()
fig.savefig(OUTPUT_DIR / "final_extra_trees_domain_importance.png", dpi=300, bbox_inches="tight")
fig.savefig(OUTPUT_DIR / "final_extra_trees_domain_importance.pdf", bbox_inches="tight")
fig.savefig(OUTPUT_DIR / "final_extra_trees_top15_raw_domain_importance.png", dpi=300, bbox_inches="tight")
fig.savefig(OUTPUT_DIR / "final_extra_trees_top15_raw_domain_importance.pdf", bbox_inches="tight")
show_or_close(fig)


def original_feature_from_encoded_name(encoded_name: str) -> str:
    name = encoded_name.split("__", 1)[-1]
    for feature in sorted(feature_columns, key=len, reverse=True):
        if name == feature or name.startswith(f"{feature}_"):
            return feature
    return name


encoded_feature_names = final_model.named_steps["preprocess"].get_feature_names_out()
encoded_importance = pd.DataFrame(
    {
        "encoded_feature": encoded_feature_names,
        "importance": final_model.named_steps["model"].feature_importances_,
    }
)
encoded_importance["feature"] = encoded_importance["encoded_feature"].apply(original_feature_from_encoded_name)
encoded_importance = encoded_importance.sort_values("importance", ascending=False).reset_index(drop=True)

feature_importance = (
    encoded_importance
    .groupby("feature", as_index=False)["importance"]
    .sum()
    .sort_values("importance", ascending=False) # type: ignore
    .reset_index(drop=True)
)
feature_importance["importance_percent"] = feature_importance["importance"] / feature_importance["importance"].sum() * 100
all_extra_trees_importance = feature_importance.merge(
    domain_importance_summary[["domain", "importance_mean", "importance_std"]].rename(
        columns={
            "domain": "feature",
            "importance_mean": "permutation_macro_f1_drop_mean",
            "importance_std": "permutation_macro_f1_drop_std",
        }
    ),
    on="feature",
    how="left",
)

encoded_importance.to_csv(OUTPUT_DIR / "final_extra_trees_encoded_feature_importance.csv", index=False)
feature_importance.to_csv(OUTPUT_DIR / "final_extra_trees_grouped_feature_importance.csv", index=False)
all_extra_trees_importance.to_csv(OUTPUT_DIR / "final_extra_trees_all_feature_importance.csv", index=False)

with pd.option_context("display.max_rows", None, "display.max_columns", None, "display.max_colwidth", None):
    display(all_extra_trees_importance)

top_extra_trees_builtin_importance = feature_importance.head(RAW_IMPORTANCE_TOP_N).copy()

plot_data = top_extra_trees_builtin_importance.sort_values("importance", ascending=True)
fig, ax = plt.subplots(figsize=(7.2, 6.2))
ax.barh(plot_data["feature"], plot_data["importance_percent"], color=PLOT_COLORS["final_model"], edgecolor="#6B7280", linewidth=0.45)
ax.set_title("Final Extra Trees built-in feature importance")
ax.set_xlabel("Share of total Extra Trees impurity-based importance (%)")
ax.grid(False)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
for value, feature in zip(plot_data["importance_percent"], plot_data["feature"]):
    ax.text(value + plot_data["importance_percent"].max() * 0.015, feature, f"{value:.1f}%", va="center", fontsize=8.5, color=PLOT_COLORS["text"])
fig.tight_layout()
fig.savefig(OUTPUT_DIR / "final_extra_trees_grouped_feature_importance.png", dpi=300, bbox_inches="tight")
fig.savefig(OUTPUT_DIR / "final_extra_trees_grouped_feature_importance.pdf", bbox_inches="tight")
fig.savefig(OUTPUT_DIR / "final_extra_trees_top15_builtin_feature_importance.png", dpi=300, bbox_inches="tight")
fig.savefig(OUTPUT_DIR / "final_extra_trees_top15_builtin_feature_importance.pdf", bbox_inches="tight")
show_or_close(fig)


def normalized_importance(series: pd.Series) -> pd.Series:
    values = series.clip(lower=0)
    max_value = values.max()
    if pd.isna(max_value) or max_value == 0:
        return values * 0
    return values / max_value


sample_size = min(SHAP_SAMPLE_SIZE, len(X))
if sample_size < len(X):
    shap_splitter = StratifiedShuffleSplit(n_splits=1, train_size=sample_size, random_state=SEED)
    shap_index, _ = next(shap_splitter.split(X, y))
else:
    shap_index = np.arange(len(X))
X_shap = X.iloc[shap_index]
y_shap = y.iloc[shap_index]
shap_sample_distribution = y_shap.value_counts().reindex(CLASS_ORDER).rename_axis("class").reset_index(name="rows")

X_shap_transformed = final_model.named_steps["preprocess"].transform(X_shap)
shap_explainer = shap.TreeExplainer(final_model.named_steps["model"])
shap_values = shap_explainer.shap_values(X_shap_transformed, check_additivity=False)
if isinstance(shap_values, list):
    shap_mean_abs = np.stack([np.abs(values) for values in shap_values], axis=-1).mean(axis=(0, 2))
else:
    shap_array = np.asarray(shap_values)
    if shap_array.ndim == 3:
        shap_mean_abs = np.abs(shap_array).mean(axis=(0, 2))
    elif shap_array.ndim == 2:
        shap_mean_abs = np.abs(shap_array).mean(axis=0)
    else:
        raise ValueError(f"Unsupported SHAP array shape: {shap_array.shape}")

encoded_shap = pd.DataFrame(
    {
        "encoded_feature": encoded_feature_names,
        "shap_mean_abs": shap_mean_abs,
    }
)
encoded_shap["feature"] = encoded_shap["encoded_feature"].apply(original_feature_from_encoded_name)
shap_importance = (
    encoded_shap
    .groupby("feature", as_index=False)["shap_mean_abs"]
    .sum()
    .sort_values("shap_mean_abs", ascending=False) # type: ignore
    .reset_index(drop=True)
)
shap_importance["shap_importance_percent"] = shap_importance["shap_mean_abs"] / shap_importance["shap_mean_abs"].sum() * 100

permutation_importance = domain_importance_summary.rename(
    columns={
        "domain": "feature",
        "importance_mean": "permutation_macro_f1_drop_mean",
        "importance_std": "permutation_macro_f1_drop_std",
    }
)[["feature", "permutation_macro_f1_drop_mean", "permutation_macro_f1_drop_std"]]

importance_comparison = (
    pd.DataFrame({"feature": feature_columns})
    .merge(permutation_importance, on="feature", how="left")
    .merge(
        feature_importance.rename(
            columns={
                "importance": "builtin_importance",
                "importance_percent": "builtin_importance_percent",
            }
        ),
        on="feature",
        how="left",
    )
    .merge(shap_importance, on="feature", how="left")
)
value_columns = [
    "permutation_macro_f1_drop_mean",
    "builtin_importance",
    "builtin_importance_percent",
    "shap_mean_abs",
    "shap_importance_percent",
]
importance_comparison[value_columns] = importance_comparison[value_columns].fillna(0)
importance_comparison["permutation_normalized"] = normalized_importance(importance_comparison["permutation_macro_f1_drop_mean"])
importance_comparison["builtin_normalized"] = normalized_importance(importance_comparison["builtin_importance"])
importance_comparison["shap_normalized"] = normalized_importance(importance_comparison["shap_mean_abs"])
importance_comparison["mean_normalized_importance"] = importance_comparison[
    ["permutation_normalized", "builtin_normalized", "shap_normalized"]
].mean(axis=1)
importance_comparison["shap_sample_rows"] = sample_size
importance_comparison["shap_sample_note"] = "stratified deterministic sample from full dataset"
importance_comparison = importance_comparison.sort_values("builtin_importance", ascending=False).reset_index(drop=True)

importance_comparison.to_csv(OUTPUT_DIR / "final_extra_trees_importance_method_comparison.csv", index=False)
shap_sample_distribution.to_csv(OUTPUT_DIR / "final_extra_trees_shap_sample_distribution.csv", index=False)
encoded_shap.to_csv(OUTPUT_DIR / "final_extra_trees_encoded_shap_importance.csv", index=False)
shap_importance.to_csv(OUTPUT_DIR / "final_extra_trees_grouped_shap_importance.csv", index=False)

with pd.option_context("display.max_rows", None, "display.max_columns", None, "display.max_colwidth", None):
    display(importance_comparison)

plot_data = importance_comparison.head(RAW_IMPORTANCE_TOP_N).sort_values("builtin_importance", ascending=True)
method_panels = [
    (
        "builtin_normalized",
        "builtin_importance_percent",
        "Built-in Extra Trees\nimpurity importance",
        lambda value: f"{value:.1f}%",
        "#8AA1B4",
    ),
    (
        "permutation_normalized",
        "permutation_macro_f1_drop_mean",
        "Permutation\nheld-out macro F1 drop",
        lambda value: f"{value:.3f}",
        PLOT_COLORS["final_model"],
    ),
    (
        "shap_normalized",
        "shap_importance_percent",
        "TreeSHAP\nmean |SHAP|",
        lambda value: f"{value:.1f}%",
        "#D4A72C",
    ),
]

fig, axes = plt.subplots(ncols=3, figsize=(13.2, 6.4), sharey=True)
for axis_index, (ax, (normalized_column, raw_column, title, formatter, color)) in enumerate(zip(axes, method_panels)):
    ax.barh(
        plot_data["feature"],
        plot_data[normalized_column],
        color=color,
        edgecolor="#6B7280",
        linewidth=0.45,
        alpha=0.94,
    )
    ax.set_title(title)
    ax.set_xlim(0, 1.12)
    ax.set_xlabel("Normalized importance")
    ax.grid(axis="x", color=PLOT_COLORS["grid"], linewidth=0.7)
    ax.grid(axis="y", visible=False)
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    if axis_index > 0:
        ax.tick_params(axis="y", left=False, labelleft=False)
        ax.spines["left"].set_visible(False)
    else:
        ax.tick_params(axis="y", length=0)
    for feature, normalized_value, raw_value in zip(plot_data["feature"], plot_data[normalized_column], plot_data[raw_column]):
        ax.text(
            min(normalized_value + 0.025, 1.1),
            feature,
            formatter(raw_value),
            va="center",
            fontsize=8.3,
            color=PLOT_COLORS["text"],
        )

fig.suptitle("Final Extra Trees feature-importance comparison", y=1.02, fontsize=13, fontweight="bold")
fig.text(
    0.5,
    -0.02,
    f"Rows show the top predictors ranked by built-in Extra Trees importance. TreeSHAP uses a stratified sample of {sample_size:,} rows; labels show each method's original scale.",
    ha="center",
    color=PLOT_COLORS["text"],
    fontsize=8.8,
)
fig.tight_layout()
fig.savefig(OUTPUT_DIR / "final_extra_trees_importance_method_comparison.png", dpi=300, bbox_inches="tight")
fig.savefig(OUTPUT_DIR / "final_extra_trees_importance_method_comparison.pdf", bbox_inches="tight")
fig.savefig(OUTPUT_DIR / "final_extra_trees_top15_importance_method_comparison.png", dpi=300, bbox_inches="tight")
fig.savefig(OUTPUT_DIR / "final_extra_trees_top15_importance_method_comparison.pdf", bbox_inches="tight")
show_or_close(fig)
```
