# IEQ Final Model Comparison

This notebook trains and evaluates all fixed paper-facing IEQ model families using the model-ready dataset. It does not clean data, impute missing values, or search for new hyperparameters. Each model uses the selected hyperparameters from the previous tuning work.

## 1. Imports and Reproducibility Settings

```python
from __future__ import annotations

import platform
import sys
import time
import warnings
from pathlib import Path

import matplotlib
if "ipykernel" not in sys.modules:
    matplotlib.use("Agg")
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn
from sklearn.base import clone
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import ExtraTreesClassifier, GradientBoostingClassifier, RandomForestClassifier
from sklearn.exceptions import ConvergenceWarning
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, balanced_accuracy_score, confusion_matrix, f1_score, mean_absolute_error
from sklearn.model_selection import StratifiedKFold
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

try:
    from IPython.display import display # type: ignore
except ModuleNotFoundError:
    def display(value):
        print(value)

warnings.filterwarnings("ignore", category=ConvergenceWarning)

SEED = 20260507
N_SPLITS = 5
TARGET = "IEQ satisfaction 3-class"
DROP_FROM_FEATURES = [TARGET, "TimeVote"]
CLASS_ORDER = ["dissatisfied", "neutral", "satisfied"]
CLASS_TO_NUMBER = {label: index for index, label in enumerate(CLASS_ORDER)}
PERFORMANCE_METRICS = ["accuracy", "balanced_accuracy", "macro_f1", "weighted_f1", "ordinal_mae"]
DOMAIN_IMPORTANCE_REPEATS = 5
RAW_IMPORTANCE_TOP_N = 15
FEATURE_CATEGORIES = {
    "Thermal environment": ["Temperature", "RH", "CLO", "Ttrend", "AC on/off"],
    "Indoor air quality": ["CO2"],
    "Lighting": ["Lighting"],
    "Acoustic": ["Sound"],
    "Occupant / schedule": ["Student", "Age", "Gender", "EA", "Vote hour", "Vote weekday", "Moment"],
    "Classroom position": ["LocationBack", "LocationFront", "LocationLeft", "LocationRight", "LocationMiddle"],
}
PLOT_COLORS = {
    "selected_model": "#2A9D8F",
    "reference_model": "#B8C0C8",
    "text": "#111827",
    "grid": "#D9DEE3",
}

plt.rcParams.update(
    {
        "figure.facecolor": "white",
        "axes.facecolor": "white",
        "savefig.facecolor": "white",
        "font.family": "sans-serif",
        "font.size": 10.5,
        "axes.edgecolor": "#AAB2B8",
        "axes.grid": True,
        "grid.color": PLOT_COLORS["grid"],
        "grid.alpha": 0.85,
    }
)


def show_or_close(fig) -> None:
    if "ipykernel" in sys.modules:
        plt.show()
    else:
        plt.close(fig)

MODEL_ORDER = ["Logistic regression", "Random Forest", "Extra Trees", "Gradient boosting", "ANN / MLP"]
SELECTED_MODEL = "Extra Trees"

FINAL_MODEL_CONFIGS = {
    "Logistic regression": {
        "estimator": LogisticRegression(C=30.0, class_weight="balanced", max_iter=2500, random_state=SEED),
        "selected_params": {"C": 30.0, "class_weight": "balanced"},
    },
    "Random Forest": {
        "estimator": RandomForestClassifier(
            n_estimators=300,
            max_depth=20,
            max_features="sqrt",
            min_samples_leaf=4,
            min_samples_split=10,
            class_weight="balanced",
            n_jobs=1,
            random_state=SEED,
        ),
        "selected_params": {
            "n_estimators": 300,
            "max_depth": 20,
            "max_features": "sqrt",
            "min_samples_leaf": 4,
            "min_samples_split": 10,
            "class_weight": "balanced",
        },
    },
    "Extra Trees": {
        "estimator": ExtraTreesClassifier(
            n_estimators=700,
            criterion="entropy",
            bootstrap=True,
            max_samples=0.9,
            max_features=None,
            max_depth=20,
            min_samples_leaf=2,
            min_samples_split=10,
            class_weight="balanced",
            n_jobs=1,
            random_state=SEED,
        ),
        "selected_params": {
            "n_estimators": 700,
            "criterion": "entropy",
            "bootstrap": True,
            "max_samples": 0.9,
            "max_features": None,
            "max_depth": 20,
            "min_samples_leaf": 2,
            "min_samples_split": 10,
            "class_weight": "balanced",
        },
    },
    "Gradient boosting": {
        "estimator": GradientBoostingClassifier(
            n_estimators=120,
            learning_rate=0.08,
            max_depth=3,
            max_features="sqrt",
            min_samples_leaf=3,
            subsample=1.0,
            random_state=SEED,
        ),
        "selected_params": {
            "n_estimators": 120,
            "learning_rate": 0.08,
            "max_depth": 3,
            "max_features": "sqrt",
            "min_samples_leaf": 3,
            "subsample": 1.0,
        },
    },
    "ANN / MLP": {
        "estimator": MLPClassifier(
            hidden_layer_sizes=(32,),
            activation="relu",
            alpha=0.01,
            learning_rate_init=0.003,
            max_iter=250,
            early_stopping=False,
            random_state=SEED,
        ),
        "selected_params": {
            "hidden_layer_sizes": (32,),
            "activation": "relu",
            "alpha": 0.01,
            "learning_rate_init": 0.003,
        },
    },
}

print("Python:", platform.python_version())
print("pandas:", pd.__version__)
print("numpy:", np.__version__)
print("scikit-learn:", sklearn.__version__)
print("Random seed:", SEED)
```

## 2. Locate and Load the Model-Ready Dataset

The notebook first looks for the original Prism folder structure. If the notebook is shared separately, placing `ieq_model_dataset.xlsx` in the same working folder is enough.

```python
DATASET_FILE_NAME = "ieq_model_dataset.xlsx"
DATASET_RELATIVE_PATH = Path("02_Datasets") / "model_ready" / DATASET_FILE_NAME


def find_dataset_path() -> Path:
    candidates = []
    for root in [Path.cwd(), *Path.cwd().parents]:
        candidates.extend(
            [
                root / DATASET_RELATIVE_PATH,
                root / "Prism" / DATASET_RELATIVE_PATH,
                root / DATASET_FILE_NAME,
            ]
        )
    for candidate in candidates:
        if candidate.exists():
            return candidate.resolve()
    raise FileNotFoundError(
        "Could not find ieq_model_dataset.xlsx. Keep the Prism folder structure, "
        "or place ieq_model_dataset.xlsx in the current working folder."
    )


DATASET_PATH = find_dataset_path()
if DATASET_PATH.parent.name == "model_ready" and DATASET_PATH.parent.parent.name == "02_Datasets":
    PROJECT_ROOT = DATASET_PATH.parent.parent.parent
    OUTPUT_DIR = PROJECT_ROOT / "03_Code" / "ieq_paper" / "01_notebook" / "ieq_all_models_outputs"
else:
    OUTPUT_DIR = Path.cwd() / "ieq_all_models_outputs"
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

data = pd.read_excel(DATASET_PATH, sheet_name="data")

print("Dataset:", DATASET_PATH)
print("Rows, columns:", data.shape)
display(data.head())
```

## 3. Select Predictors

`IEQ satisfaction 3-class` is the target and must not be used as a predictor. `TimeVote` is the original timestamp metadata column, so the model uses the derived time-context inputs `Vote hour` and `Vote weekday` instead.

```python
feature_columns = [column for column in data.columns if column not in DROP_FROM_FEATURES]

X = data[feature_columns].copy()
y = data[TARGET].astype(str)

numeric_features = X.select_dtypes(include=["number"]).columns.tolist()
categorical_features = [column for column in X.columns if column not in numeric_features]

print("Target:", TARGET)
print("Number of predictors:", len(feature_columns))
print("Numeric predictors:", numeric_features)
print("Categorical / boolean predictors:", categorical_features)
print("Class distribution:")
display(y.value_counts().reindex(CLASS_ORDER).rename("rows").to_frame())

category_feature_map = {
    category: [feature for feature in features if feature in feature_columns]
    for category, features in FEATURE_CATEGORIES.items()
}
category_feature_map = {category: features for category, features in category_feature_map.items() if features}
domain_feature_map = {feature: [feature] for feature in feature_columns}
domain_feature_table = pd.DataFrame(
    [
        {"domain": category, "n_features": len(features), "features": ", ".join(features)}
        for category, features in category_feature_map.items()
    ]
)
with pd.option_context("display.max_colwidth", None, "display.max_rows", None):
    display(domain_feature_table)
```

## 4. Fixed Model Pipelines

The preprocessing is fitted only inside each training fold during cross-validation. Numeric predictors are standardized and categorical or boolean predictors are one-hot encoded. Each classifier uses fixed hyperparameters from the previous tuning work.

```python
def make_one_hot_encoder():
    try:
        return OneHotEncoder(handle_unknown="ignore", sparse_output=False)
    except TypeError:
        return OneHotEncoder(handle_unknown="ignore", sparse=False)


def make_preprocessor():
    return ColumnTransformer(
        transformers=[
            ("numeric", StandardScaler(), numeric_features),
            ("categorical", make_one_hot_encoder(), categorical_features),
        ],
        remainder="drop",
    )


def make_model(model_name: str):
    return clone(FINAL_MODEL_CONFIGS[model_name]["estimator"])


def make_pipeline(model_name: str):
    return Pipeline(
        steps=[
            ("preprocess", make_preprocessor()),
            ("model", make_model(model_name)),
        ]
    )


model_parameter_rows = []
for model_name in MODEL_ORDER:
    for parameter, value in FINAL_MODEL_CONFIGS[model_name]["selected_params"].items():
        model_parameter_rows.append(
            {
                "model": model_name,
                "parameter": parameter,
                "value": value,
            }
        )

model_parameters = pd.DataFrame(model_parameter_rows)
display(model_parameters)
```

## 5. Deterministic 5-Fold Cross-Validation

This section evaluates all fixed model configurations. No hyperparameter search is performed.

```python
def ordinal_mae(y_true, y_pred):
    true_numbers = pd.Series(y_true).map(CLASS_TO_NUMBER).to_numpy()
    pred_numbers = pd.Series(y_pred).map(CLASS_TO_NUMBER).to_numpy()
    return float(mean_absolute_error(true_numbers, pred_numbers))


def evaluate_predictions(y_true, y_pred):
    return {
        "accuracy": float(accuracy_score(y_true, y_pred)),
        "balanced_accuracy": float(balanced_accuracy_score(y_true, y_pred)),
        "macro_f1": float(f1_score(y_true, y_pred, average="macro", zero_division=0)),
        "weighted_f1": float(f1_score(y_true, y_pred, average="weighted", zero_division=0)),
        "ordinal_mae": float(ordinal_mae(y_true, y_pred)),
    }


def grouped_domain_permutation_rows(pipeline, X_test, y_test, baseline_macro_f1, model_name, fold):
    rows = []
    model_offset = MODEL_ORDER.index(model_name) + 1
    rng = np.random.default_rng(SEED + model_offset * 10000 + fold * 1000)
    for domain, domain_features in domain_feature_map.items():
        feature_list = ", ".join(domain_features)
        for repeat in range(1, DOMAIN_IMPORTANCE_REPEATS + 1):
            shuffled_positions = rng.permutation(len(X_test))
            X_permuted = X_test.copy()
            for feature in domain_features:
                X_permuted[feature] = X_permuted[feature].iloc[shuffled_positions].to_numpy()
            permuted_pred = pipeline.predict(X_permuted)
            permuted_macro_f1 = float(f1_score(y_test, permuted_pred, average="macro", zero_division=0))
            rows.append(
                {
                    "model": model_name,
                    "fold": fold,
                    "repeat": repeat,
                    "domain": domain,
                    "features": feature_list,
                    "baseline_macro_f1": baseline_macro_f1,
                    "permuted_macro_f1": permuted_macro_f1,
                    "macro_f1_drop": baseline_macro_f1 - permuted_macro_f1,
                }
            )
    return rows


cv = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=SEED)
fold_rows = []
prediction_rows = []
domain_importance_rows = []

for model_name in MODEL_ORDER:
    print(f"Evaluating {model_name}...")
    for fold, (train_index, test_index) in enumerate(cv.split(X, y), start=1):
        pipeline = make_pipeline(model_name)
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        start = time.perf_counter()
        pipeline.fit(X_train, y_train)
        fit_seconds = time.perf_counter() - start

        start = time.perf_counter()
        y_pred = pipeline.predict(X_test)
        predict_seconds = time.perf_counter() - start

        metrics = evaluate_predictions(y_test, y_pred)
        domain_importance_rows.extend(grouped_domain_permutation_rows(pipeline, X_test, y_test, metrics["macro_f1"], model_name, fold))
        fold_rows.append(
            {
                "model": model_name,
                "fold": fold,
                "train_rows": len(train_index),
                "test_rows": len(test_index),
                **metrics,
                "fit_seconds": fit_seconds,
                "predict_seconds": predict_seconds,
                "total_seconds": fit_seconds + predict_seconds,
            }
        )

        for source_index, true_label, predicted_label in zip(test_index, y_test, y_pred):
            prediction_rows.append(
                {
                    "model": model_name,
                    "source_index": int(source_index),
                    "fold": fold,
                    "y_true": true_label,
                    "y_pred": predicted_label,
                }
            )

fold_results = pd.DataFrame(fold_rows)
cv_predictions = pd.DataFrame(prediction_rows)
domain_importance = pd.DataFrame(domain_importance_rows)
domain_importance_summary = (
    domain_importance
    .groupby(["model", "domain", "features"], as_index=False)
    .agg(
        importance_mean=("macro_f1_drop", "mean"),
        importance_std=("macro_f1_drop", "std"),
        baseline_macro_f1_mean=("baseline_macro_f1", "mean"),
        permuted_macro_f1_mean=("permuted_macro_f1", "mean"),
    )
)
domain_importance_summary["model"] = pd.Categorical(domain_importance_summary["model"], categories=MODEL_ORDER, ordered=True)
domain_importance_summary = domain_importance_summary.sort_values(["model", "importance_mean"], ascending=[True, False]).reset_index(drop=True)

summary = fold_results.groupby("model")[PERFORMANCE_METRICS].agg(["mean", "std"])
summary.columns = [f"{metric}_{stat}" for metric, stat in summary.columns]
summary = summary.reset_index()
summary["model"] = pd.Categorical(summary["model"], categories=MODEL_ORDER, ordered=True)
summary = summary.sort_values("model").reset_index(drop=True)
fold_metrics = fold_results[["model", "fold", "train_rows", "test_rows", *PERFORMANCE_METRICS]].copy()

display(summary.sort_values("macro_f1_mean", ascending=False))
```

## 6. Confusion Matrices and Saved Outputs

```python
confusion_tables = {}
confusion_rows = []
for model_name in MODEL_ORDER:
    model_predictions = cv_predictions.loc[cv_predictions["model"].eq(model_name)]
    confusion = pd.DataFrame(
        confusion_matrix(model_predictions["y_true"], model_predictions["y_pred"], labels=CLASS_ORDER),
        index=[f"true_{label}" for label in CLASS_ORDER],
        columns=[f"pred_{label}" for label in CLASS_ORDER],
    )
    confusion_tables[model_name] = confusion
    for true_label in CLASS_ORDER:
        for pred_label in CLASS_ORDER:
            confusion_rows.append(
                {
                    "model": model_name,
                    "true_class": true_label,
                    "predicted_class": pred_label,
                    "count": int(confusion.loc[f"true_{true_label}", f"pred_{pred_label}"]),
                }
            )

confusion_long = pd.DataFrame(confusion_rows)

fold_metrics.to_csv(OUTPUT_DIR / "all_models_5fold_folds.csv", index=False)
summary.to_csv(OUTPUT_DIR / "all_models_5fold_summary.csv", index=False)
cv_predictions.to_csv(OUTPUT_DIR / "all_models_5fold_predictions.csv", index=False)
confusion_long.to_csv(OUTPUT_DIR / "all_models_confusion_matrices.csv", index=False)
model_parameters.to_csv(OUTPUT_DIR / "all_models_fixed_hyperparameters.csv", index=False)
domain_importance.to_csv(OUTPUT_DIR / "all_models_domain_importance_folds.csv", index=False)
domain_importance_summary.to_csv(OUTPUT_DIR / "all_models_domain_importance_summary.csv", index=False)

display(confusion_tables[SELECTED_MODEL])

fig, axes = plt.subplots(nrows=1, ncols=len(MODEL_ORDER), figsize=(18.0, 4.0), sharex=True, sharey=True)
for ax, model_name in zip(axes, MODEL_ORDER):
    confusion = confusion_tables[model_name]
    confusion_percent = confusion.div(confusion.sum(axis=1), axis=0) * 100
    image = ax.imshow(confusion_percent, cmap="Blues", vmin=0, vmax=100)
    ax.set_title(model_name)
    ax.set_xticks(range(len(CLASS_ORDER)))
    ax.set_xticklabels(CLASS_ORDER, rotation=35, ha="right")
    ax.set_yticks(range(len(CLASS_ORDER)))
    ax.set_yticklabels(CLASS_ORDER)
    for row_index in range(len(CLASS_ORDER)):
        for column_index in range(len(CLASS_ORDER)):
            count = confusion.iloc[row_index, column_index]
            percent = confusion_percent.iloc[row_index, column_index]
            text_color = "white" if percent >= 50 else PLOT_COLORS["text"]
            ax.text(column_index, row_index, f"{count}\n{percent:.1f}%", ha="center", va="center", fontsize=8, color=text_color)

axes[0].set_ylabel("True class")
for ax in axes:
    ax.set_xlabel("Predicted class")
colorbar = fig.colorbar(image, ax=axes.ravel().tolist(), fraction=0.018, pad=0.02)
colorbar.set_label("Row percentage")
fig.suptitle("Confusion matrices by model", y=1.04, fontsize=13, fontweight="bold")
fig.savefig(OUTPUT_DIR / "all_models_confusion_matrices.png", dpi=300, bbox_inches="tight")
fig.savefig(OUTPUT_DIR / "all_models_confusion_matrices.pdf", bbox_inches="tight")
show_or_close(fig)

print("Saved outputs to:", OUTPUT_DIR.resolve())
```

## 7. Fixed Model Comparison

All values in this section are computed in this notebook using the fixed model hyperparameters. No hyperparameter search is performed.

```python
comparison_results = summary[[
    "model",
    "accuracy_mean",
    "balanced_accuracy_mean",
    "macro_f1_mean",
    "weighted_f1_mean",
    "ordinal_mae_mean",
]].copy()
comparison_results = comparison_results.rename(
    columns={
        "accuracy_mean": "accuracy",
        "balanced_accuracy_mean": "balanced_accuracy",
        "macro_f1_mean": "macro_f1",
        "weighted_f1_mean": "weighted_f1",
        "ordinal_mae_mean": "ordinal_mae",
    }
)

display(comparison_results)

COMPARISON_PLOT_METRICS = ["accuracy", "balanced_accuracy", "macro_f1", "ordinal_mae"]

metric_labels = {
    "accuracy": "Accuracy",
    "balanced_accuracy": "Balanced\naccuracy",
    "macro_f1": "Macro F1",
    "weighted_f1": "Weighted F1",
    "ordinal_mae": "Ordinal MAE\n(lower is better)",
}

model_labels = comparison_results["model"].tolist()
x_positions = np.arange(len(model_labels))
bar_colors = [
    PLOT_COLORS["selected_model"] if model == SELECTED_MODEL else PLOT_COLORS["reference_model"]
    for model in model_labels
]

fig, axes = plt.subplots(ncols=len(COMPARISON_PLOT_METRICS), figsize=(14.0, 5.2))
for ax, metric in zip(axes, COMPARISON_PLOT_METRICS):
    values = comparison_results[metric].astype(float).to_numpy()
    ax.bar(x_positions, values, color=bar_colors, edgecolor="#6B7280", linewidth=0.45)
    ax.set_title(metric_labels[metric])
    ax.set_xticks(x_positions)
    ax.set_xticklabels(model_labels, rotation=35, ha="right")
    ax.grid(axis="y", color=PLOT_COLORS["grid"], alpha=0.85)
    ax.grid(axis="x", visible=False)
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_color("#AAB2B8")
    ax.spines["bottom"].set_color("#AAB2B8")
    if metric == "ordinal_mae":
        ax.set_ylim(0, max(values) * 1.18)
    else:
        ax.set_ylim(0, 1.0)
    for x_position, value in zip(x_positions, values):
        ax.text(x_position, value + ax.get_ylim()[1] * 0.018, f"{value:.3f}", ha="center", va="bottom", fontsize=8.5, color=PLOT_COLORS["text"])

fig.suptitle("Fixed-hyperparameter IEQ model comparison", y=1.03, fontsize=13, fontweight="bold")
fig.text(0.5, -0.02, "All bars are computed in this notebook; Extra Trees is highlighted because it is the selected paper-facing model.", ha="center", color=PLOT_COLORS["text"])
fig.tight_layout()
fig.savefig(OUTPUT_DIR / "all_models_comparison.png", dpi=300, bbox_inches="tight")
fig.savefig(OUTPUT_DIR / "all_models_comparison.pdf", bbox_inches="tight")
show_or_close(fig)

comparison_results.to_csv(OUTPUT_DIR / "all_models_comparison.csv", index=False)
```

## 8. Train All Fixed Models on All Rows

The cross-validation above estimates performance. This cell trains each fixed pipeline on all available model-ready rows, creating fitted model objects for downstream prediction or export.

```python
final_models = {}
final_fit_rows = []

for model_name in MODEL_ORDER:
    pipeline = make_pipeline(model_name)
    start = time.perf_counter()
    pipeline.fit(X, y)
    fit_seconds = time.perf_counter() - start
    final_models[model_name] = pipeline
    final_fit_rows.append(
        {
            "model": model_name,
            "training_rows": len(X),
            "predictors": len(feature_columns),
            "fit_seconds": fit_seconds,
        }
    )

final_fit_summary = pd.DataFrame(final_fit_rows)
display(final_fit_summary)
print("Trained final fixed models on all rows:", ", ".join(final_models.keys()))
print("Training rows:", len(X))
print("Predictors:", len(feature_columns))
final_models[SELECTED_MODEL]
```

## 9. Raw Domain Importance

Raw-domain permutation importance is computed with the same stratified 5-fold cross-validation as the fixed-hyperparameter model comparison. For each trained fold model, the baseline held-out macro F1 is computed first. Then each model-ready predictor is shuffled within the held-out fold, all other predictors are left unchanged, and macro F1 is recomputed. This permutation is repeated five times per fold. Importance is the mean held-out macro-F1 decrease after shuffling; larger positive values indicate predictors the model relied on more. The plotted bars show these means only; fold/repeat variability remains available in the CSV output but is not drawn in the figure.

```python
domain_importance.to_csv(OUTPUT_DIR / "all_models_domain_importance_folds.csv", index=False)
domain_importance_summary.to_csv(OUTPUT_DIR / "all_models_domain_importance_summary.csv", index=False)
top_raw_domain_importance = domain_importance_summary.groupby("model", sort=False).head(RAW_IMPORTANCE_TOP_N).copy()
print("Saved", OUTPUT_DIR / "all_models_domain_importance_folds.csv")
print("Saved", OUTPUT_DIR / "all_models_domain_importance_summary.csv")
```

```python
fig, axes = plt.subplots(nrows=len(MODEL_ORDER), ncols=1, figsize=(8.6, 15.5), sharex=True)

for ax, model_name in zip(axes, MODEL_ORDER):
    model_domains = (
        domain_importance_summary[domain_importance_summary["model"] == model_name]
        .head(RAW_IMPORTANCE_TOP_N)
        .sort_values("importance_mean", ascending=True)
    )
    color = PLOT_COLORS["selected_model"] if model_name == SELECTED_MODEL else PLOT_COLORS["reference_model"]
    ax.barh(model_domains["domain"], model_domains["importance_mean"], color=color, alpha=0.9)
    ax.set_title(model_name, loc="left", fontsize=10, fontweight="bold", color=PLOT_COLORS["text"])
    ax.set_xlabel("Mean held-out macro F1 decrease after permutation")
    ax.grid(axis="x", color=PLOT_COLORS["grid"], linewidth=0.7)
    for spine in ["top", "right", "left"]:
        ax.spines[spine].set_visible(False)
    ax.tick_params(axis="y", length=0, labelsize=8)

fig.suptitle("Top 15 raw-domain importance by fixed IEQ model", y=1.01, fontsize=13, fontweight="bold")
fig.text(0.5, -0.01, "Bars show mean macro-F1 drop after shuffling one model-ready predictor in held-out CV folds; higher values indicate stronger model reliance.", ha="center", color=PLOT_COLORS["text"])
fig.tight_layout()
fig.savefig(OUTPUT_DIR / "all_models_domain_importance.png", dpi=300, bbox_inches="tight")
fig.savefig(OUTPUT_DIR / "all_models_domain_importance.pdf", bbox_inches="tight")
fig.savefig(OUTPUT_DIR / "all_models_top15_raw_domain_importance.png", dpi=300, bbox_inches="tight")
fig.savefig(OUTPUT_DIR / "all_models_top15_raw_domain_importance.pdf", bbox_inches="tight")
show_or_close(fig)
```
